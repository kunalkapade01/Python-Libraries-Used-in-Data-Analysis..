- ùêíùêúùêöùê≠ùê≠ùêûùê´ ùêèùê•ùê®ùê≠ :

```python
#ploting using matplotlib

#plt scatter

plt.scatter(x,y,c='g') 
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Graph in 2D')
plt.savefig('Test.png')
```

```python
y=x*x     #curved shape diagram:
```

```python
# plt plot

plt.plot(x,y,'r*', linestyle='dashed', linewidth=2, markersize=12)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('2D Diagram')
```

```python
#create subplot

plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(x,y,'g*--')
plt.subplot(2,2,3)
plt.plot(x,y,'bo')
plt.subplot(2,2,4)
plt.plot(x,y,'go')
```

```python
x=np.arange(1,11)
y = 3 * x + 5
plt.title('Matplot demo')
plt.xlabel('x-axis caption')
plt.ylabel('y-axis caption')
plt.plot(x,y,'g')
```

- ùêÅùêöùê´ ùêèùê•ùê®ùê≠ :

```python
# bar plot

x=[1,2,3,4]
y=[3,4,6,7]

x2=[4,7,3,6]
y2=[2,4,6,8]

plt.bar(x,y,)
plt.bar(x2,y2,color ='g')
plt.title('Bar graph')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
```

- ùêÅùê®ùê± ùêèùê•ùê®ùê≠ :

```python
# Box plot:

data = [np.random.normal(0, std, 100) for std in range(1,4)]

#Rectangular Box plot
plt.boxplot(data)       #vert=True, patch_artist=False
plt.show()
```

- ùêèùê¢ ùêÇùê°ùêöùê´ùê≠ :

```python
#data to plot
labels = "python","power-Bi","sql","AI"
sizes = [215,200,195,255]
colors = ['yellow','gold','skyblue','green']
explode = (0,0,0,0.5) #explode to data

#plot
plt.pie(sizes, explode, labels, colors, autopct='%1.1f%%', shadow=False)
plt.axis('equal')
plt.show()
```
