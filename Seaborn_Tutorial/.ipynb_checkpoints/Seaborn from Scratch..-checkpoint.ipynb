# Seaborn_Tutorial: To Visualize this data use seaborn 

```python
# Import library :
import seaborn as sns
```

```python
# Load iris datasets :
iris = sns.load_dataset('iris')
```

```python
iris
```

```python
iris.corr()
```

```python
sns.distplot(iris['sepal_length'], kde=True)
```

```python
sns.pairplot(iris, hue='species')
```

```python
sns.barplot(x='species', y='sepal_length', data=iris)
```

# dataset: tips

```python
# Load Tips datasets :
tips = sns.load_dataset('tips')
```

```python
tips
```

- 𝐁𝐚𝐫𝐩𝐥𝐨𝐭 :

```python
sns.barplot(x='sex', y='total_bill', data=tips )
```

```python
sns.barplot(x='tip', y='total_bill', data=tips)
```

```python
sns.barplot(x='smoker', y='total_bill', data=tips)
```

- 𝐜𝐨𝐮𝐧𝐭𝐩𝐥𝐨𝐭 :

```python
sns.countplot(x='sex', data=tips)
```

- 𝐛𝐨𝐱𝐩𝐥𝐨𝐭: 𝐔𝐬𝐞 𝐟𝐨𝐫 𝐭𝐨 𝐜𝐡𝐞𝐜𝐤 𝐨𝐮𝐭𝐥𝐢𝐞𝐫𝐬.

```python
sns.boxplot(x='day', y='total_bill', data=tips, hue='smoker')
```

- 𝐯𝐢𝐨𝐥𝐢𝐧𝐩𝐥𝐨𝐭:

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='sex', split=True)
```

- 𝐃𝐢𝐬𝐭𝐫𝐢𝐛𝐮𝐭𝐢𝐨𝐧 𝐏𝐥𝐨𝐭: 𝐝𝐢𝐬𝐭𝐩𝐥𝐨𝐭

```python
sns.distplot(tips['total_bill'], kde=True, bins=30)             
```

- 𝐤𝐝𝐞𝐩𝐥𝐨𝐭:

```python
sns.kdeplot(tips['total_bill'])
```

- 𝐣𝐨𝐢𝐧𝐭𝐩𝐥𝐨𝐭: 𝐬𝐜𝐚𝐭𝐭𝐞𝐫 𝐚𝐧𝐝 𝐡𝐢𝐬𝐭𝐨𝐠𝐫𝐚𝐦 𝐤𝐚 𝐣𝐨𝐢𝐧 𝐝𝐢𝐤𝐡𝐚𝐭𝐚 𝐡𝐞𝐲𝐲

```python
#kind='scatter', kind='hex', kind='reg'
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips,  kind='reg')
```

- 𝐏𝐚𝐢𝐫𝐩𝐥𝐨𝐭 :

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips, hue='sex')
```
