```python
# Import neceassary library :
import numpy as np
import pandas as pd
```

```python
x = np.random.randint(1, 99, 100)
y = np.random.randint(1, 99, 100)
z = np.random.binomial(6, 0.666, 100)
```

```python
df = pd.DataFrame({"X":x, "Y":y, "Z":z})
```

```python
df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8,13))
sns.jointplot(df, x="X", y="Y", hue="Z", palette="mako", xlim=100, ylim=100)
```
